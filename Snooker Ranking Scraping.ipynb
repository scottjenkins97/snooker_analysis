### Brief

Scrape ranking data from https://cuetracker.net/rankings/ for application to Snooker Loopy Kaggle Project.

```python
from selenium import webdriver
import pandas as pd
import numpy as np
import time
```

## Finished Scraper

This takes around 15 minutes to run on my machine.

```python
# Initialise empty lists to store data from each record in

years = [] 
first_names = []
surnames = []
world_ranks_start = []
points_start = []

driver = webdriver.Chrome(executable_path=r'C:\Users\sjenk\.ipython\chromedriver.exe')

for j in range(1981,2020):
    print('Year:',j)
    #Go to the correct URL
    url='https://cuetracker.net/rankings/{}-{}'.format(j,j+1)
    driver.get(url)
    
    #Find out: Number of records to scrape from text underneath table
    rankCount = driver.find_elements_by_xpath('/html/body/div[2]/div/div/div/div[2]/div[4]/div/div[2]/div/div[3]/div[1]/div') # The number of records in the table.
    
    #Calculate the number of pages to scrape, given that there are 50 records per page
    for dummy in rankCount:
        rankCount = int(dummy.text.split(' ')[-2])
    pageCount = int(np.ceil(rankCount/50)) 
    print('Number of pages to scrape this year:', pageCount)
    
    #Calculate the index of the 'next' button in the pagination table. This will be an input below.
    next_list_index = min(pageCount + 2,9)
    print('Next List Index:',next_list_index)
    
    #Scrape Rankings from first page
    rowCount = len(driver.find_elements_by_xpath('/html/body/div[2]/div/div/div/div[2]/div[4]/div/div[2]/div/div[2]/div/table/tbody/tr'))
    # Loop over Rows, using string splitting to separate values. Add to the respective lists.
    for i in range(1,rowCount+1):
        player = driver.find_elements_by_xpath('/html/body/div[2]/div/div/div/div[2]/div[4]/div/div[2]/div/div[2]/div/table/tbody/tr[{}]'.format(i))
        for x in player:
            details = x.text.split(' ')
            print(details)
            years.append(j)
            first_names.append(details[0])
            surnames.append(details[1])
            world_ranks_start.append(details[2])
            points_start.append(details[3])
    
    # Click the 'next' button, scrape the page, and repeat until all pages complete
    for k in range(pageCount-1):
        driver.find_element_by_xpath("/html/body/div[2]/div/div/div/div[2]/div[4]/div/div[2]/div/div[3]/div[2]/div/ul/li[{}]/a".format(next_list_index)).click()
        rowCount = len(driver.find_elements_by_xpath('/html/body/div[2]/div/div/div/div[2]/div[4]/div/div[2]/div/div[2]/div/table/tbody/tr'))
        for i in range(1,rowCount+1):
            player = driver.find_elements_by_xpath('/html/body/div[2]/div/div/div/div[2]/div[4]/div/div[2]/div/div[2]/div/table/tbody/tr[{}]'.format(i))
            for x in player:
                details = x.text.split(' ')
                print(details)
                years.append(j)
                first_names.append(details[0])
                surnames.append(details[1])
                world_ranks_start.append(details[2])
                points_start.append(details[3])
        # Page scraped, ready to click 'next' again after a brief pause      
        time.sleep(1)
        
print('Complete')
```

```python
# Compile into a DataFrame
dict = {'Year':years,
        'First_Name':first_names,
        'Surnames':surnames,
        'Ranking': world_ranks_start,
        'Points': points_start
       }
ranking_data = pd.DataFrame(dict)

ranking_data
```
